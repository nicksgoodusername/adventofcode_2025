```python
def read_input(fname: str):
    operations = []
    with open(fname, "r") as input:
        for line in input:
            operations.append(line.strip())
    return operations
```

```python
### Parts 1 & 2 

test_input = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82"""

# input = test_input.split("\n")

input = read_input("day_1.txt")

stat = 50
zero_count = 0
zero_clicks = 0

for op in input:
    mag = int(op[1:])
    direction = 1
    if op[0] == "L":
        direction = -1
    
    if stat == 0 and direction == -1:
        zero_clicks -= 1

    stat += (mag * direction)

    zero_clicks += abs(stat // 100)

    stat = stat % 100
    if stat < 0:
        stat = 100 + stat
    if stat == 0:
        zero_count += 1
        if direction == -1:
            zero_clicks += 1

print("Part 1:", zero_count)
print("Part 2:", zero_clicks)
```
