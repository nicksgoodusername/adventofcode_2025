```python
def read_input(fname: str):
    banks = []
    with open(fname, "r") as input:
        for line in input:
            banks.append(line.strip())
    return banks
input = read_input("day_3.txt")
```

```python
### Part 1

def part_1(bank_str) -> int:
    int_list = [int(x) for x in bank_str]
    max_dig_in = int_list[:-1].index(max(int_list[:-1]))  # Get the index of the first instance of the largest digit (except where it is the final digit)
    sec_list = int_list[max_dig_in + 1:]  # Get the index of the first largest digit in a list following the index of the first digit
    sec_dig_in = sec_list.index(max(sec_list))
    out_num = int(str(int_list[max_dig_in]) + str(sec_list[sec_dig_in]))
    return out_num

part_1_sum = 0
for bank in input:
    part_1_sum += part_1(bank)

print("Part 1: ", part_1_sum)
```

```python
### Part 2

def biggest_available(int_list: list, start_index: int, required_space: int) -> int:
    # Find the index of the largest digit in a list, given that it must come after a certain index 
    # and must leave some space before the end of the list
    reduced_int_list = int_list[start_index:-required_space]
    if required_space == 0:
        reduced_int_list = int_list[start_index:]
    max_dig_in_red = reduced_int_list.index(max(reduced_int_list))  # Find the index of the digit in the "reduced" list
    max_dig_in = max_dig_in_red + start_index  # Convert to absolute index for the whole list
    return max_dig_in

def part_2(bank_str: str, num_len: int = 12) -> int:
    int_list = [int(x) for x in bank_str]
    counter = num_len
    start_in = 0
    out_num = ""
    while counter > 0:
        next_dig_in = biggest_available(int_list, start_in, counter-1)
        next_dig = int_list[next_dig_in]
        out_num += str(next_dig)
        start_in = next_dig_in + 1
        counter -= 1
    return int(out_num)

part_2_sum = 0
for bank in input:
    part_2_sum += part_2(bank)

print("Part 2: ", part_2_sum)
```
