```python
import re

with open("day_2.txt") as f:
    for line in f:
        input = line.strip()

input = input.split(',')

parsed_input = []
for ivl in input:
    startstop = re.findall(r"\d+", ivl)
    parsed_input.append(startstop)
```

```python
### Parts 1 & 2

from itertools import groupby

def num_val_1(number: int) -> bool:
    num = str(number)
    N = len(num)
    if N % 2 != 0:
        return False  # Part 1 cannot work on odd-digit numbers
    i = int(N / 2)
    a = num[0:i]
    b = num[i:]
    if a == b:
        return True
    else:
        return False
    
def all_equal(chunk: list):
    g = groupby(chunk)
    return next(g, True) and not next(g, False)
    # groupby returns the number of distinct "groups" in the iterator. If all elements are equal this is 1.
    # If there's at least 1 group then the first next returns True.
    # If there's more than 1 group then the iterator is not exhausted by the second next, causing the overall statement to be False.
    # If there's only 1 group then the iterator is exhausted by the 2nd next, so overall returns True. 

def num_val_2(number: int) -> bool:
    num = str(number)
    N = len(num)

    for i in range(1, int(N/2) + 1):
        chunks = [num[x:x+i] for x in range(0, N, i)]
        if all_equal(chunks):
            return True
    return False

invalid_nums_1 = []
invalid_nums_2 = []
    
for interval in parsed_input:
    for i in range(int(interval[0]), int(interval[1]) + 1):
        if num_val_1(i):
            invalid_nums_1.append(i)
        if num_val_2(i):
            invalid_nums_2.append(i)

print("Part 1: ", sum(invalid_nums_1))
print("Part 2: ", sum(invalid_nums_2))
```
